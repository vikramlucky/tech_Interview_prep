```python
def find_string_anagrams(str, pattern):
    
    result_indexes = []
    anagrams = {}
    left = 0
    
    for ch in pattern:
        if ch not in anagrams:
            anagrams[ch] = 0
        anagrams[ch] += 1

    matched = 0 

    for right in range(len(str)):
        ch = str[right]
        
        
        if ch in anagrams:
            anagrams[ch] -= 1
            if anagrams[ch] == 0:
                matched += 1

        if matched == len(anagrams):
            result_indexes.append(left)

        if right - left + 1 > len(anagrams) - 1:
            left_char = str[left]
            if left_char in anagrams:
                if anagrams[left_char] == 0:
                    matched -= 1
                anagrams[left_char] += 1
            left += 1


  # TODO: Write your code here
    return result_indexes

```

```python
print(find_string_anagrams("ppqp", "pq"))
print(find_string_anagrams("abbcabc", "abc"))
```
