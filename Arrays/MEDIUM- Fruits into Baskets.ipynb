```python
def fruits_into_baskets(fruits):
  # TODO: Write your code here
    maxLength = float("-inf")
    fruitsBucket = {}
    windowSart = 0

    for windowEnd in range(len(fruits)):
        fruit = fruits[windowEnd]

        if fruit not in fruitsBucket:
            fruitsBucket[fruit] = 0
        fruitsBucket[fruit] += 1

        while len(fruitsBucket) > 2:
            fruit = fruits[windowSart]

            fruitsBucket[fruit] -= 1
            del fruitsBucket[fruit]
            windowSart += 1

        maxLength = max(maxLength, windowEnd - windowSart + 1)

    return maxLength
```

```python
print("Maximum number of fruits: " + str(fruits_into_baskets(['A', 'B', 'C', 'A', 'C'])))
print("Maximum number of fruits: " + str(fruits_into_baskets(['A', 'B', 'C', 'B', 'B', 'C'])))
```
