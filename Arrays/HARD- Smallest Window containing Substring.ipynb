```python
def find_substring(str, pattern):
    # TODO: Write your code here
    pattern_seen = {}
    
    for ch in pattern:
        if ch not in pattern_seen:
            pattern_seen[ch] = 0
        pattern_seen[ch] += 1
    
    matched = 0
    left = 0 
    smallest = (float("inf"), None)

    for right in range(len(str)):
        ch = str[right]

        if ch in pattern_seen:
            pattern_seen[ch] -= 1
            if pattern_seen[ch] == 0:
                matched += 1

        while matched == len(pattern_seen):
            substringlength = right - left + 1
            if substringlength < smallest[0]:
                smallest = (substringlength, str[left:right+1])
            left_char = str[left]
            if left_char in pattern_seen:
                if pattern_seen[left_char] == 0:
                    matched -= 1
                pattern_seen[left_char] += 1
            left += 1

    return smallest[1] if smallest[1] else ""
```

```python
print(find_substring("aabdec", "abc"))
print(find_substring("abdbca", "abc"))
print(find_substring("adcad", "abc"))
```
