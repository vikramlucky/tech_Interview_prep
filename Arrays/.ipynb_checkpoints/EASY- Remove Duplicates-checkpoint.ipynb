```python
def remove_duplicates(arr):
  # TODO: Write your code here
    j = 1
    for i in range(1, len(arr)):
        if arr[j-1] != arr[i]:
            arr[j] = arr[i]
            j += 1
    return arr[:j]
```

```python
print(remove_duplicates([2, 3, 3, 3, 6, 9, 9]))
print(remove_duplicates([2, 2, 2, 11]))
```
