```python
def length_of_longest_substring(str, k):
  # TODO: Write your code here
    left = 0 
    seen = {}
    longest = 0
    maxFrequencyCount = 0
    for right in range(len(str)):
        ch = str[right]
        if ch not in seen:
            seen[ch] = 0
        seen[ch] += 1
        maxFrequencyCount = max(maxFrequencyCount, seen[ch])

        while (right - left + 1 - maxFrequencyCount) > k:
            seen[str[left]] -= 1
            if seen[str[left]] == 0:
                del seen[str[left]]
            left += 1

        longest = max(longest, right - left + 1)

    return longest
```

```python

print(length_of_longest_substring("aabccbb", 2))
print(length_of_longest_substring("abbcb", 1))
print(length_of_longest_substring("abccde", 1))
```
